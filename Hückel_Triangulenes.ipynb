# Hückel Hamiltonial for Triangulenes

```python
import numpy as np                                            #importing library for calculation
import matplotlib.pyplot as plt                               #importing library for plots


# Make this a function >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
print('----------------------------------------------------------')
print('                   HÜCKEL HAMILTONIAN                     ')
print('         TCCM Winter School by Marlene Bosquez            ')
print('                                                          ')
print('                        1 -- 2                            ')
print('                       /      \                           ')
print('                      10       3 -- 4                     ')
print('                       \      /      \                    ')
print('                        11--12        5 -- 6              ')
print('                       /      \      /      \             ')
print('                      19       13--14        7 -- 8       ')
print('                       \      /      \      /      \      ')
print('                        20--21        15--16        9     ')
print('                       /      \      /      \      /      ')
print('                      26       22--23        17--18       ')
print('                       \      /      \      /             ')
print('                        27--28        24--25              ')
print('                       /      \      /                    ')
print('                      31       29--30                     ')
print('                       \      /                           ')
print('                        32--33                            ')
print('                                                          ')
print('                                                          ')
print('----------------------------------------------------------')

hex = int(input("Enter the number of hexagons per side: "))   # Ask the user about the number of hexagons per triangle's side
for L in [hex]:                                               # L is defining the side of the triangle

    lista = [[0, 0, 0, True]]                                 # To create the first entry
    for y in range(L+1):                                      # Iterate over all Ys
        for x in range(L):                                    # Iterate over all Xs
            if lista[-1][3]:                                  # If the last atom is red
                tempb = [ x + 1, y, lista[-1][2], False]      # Make the next blue atom
                tempr = [ x + 1, y, lista[-1][2] + 1, True]   # Make the next red atom
                if tempr[2] <= L:                             # If the Z coordinate is reasonable
                    lista.append( tempb )                     # Add the blue coordinates to the list
                    lista.append( tempr )                     # Add the red coordinates to the list
            else:                                             # Ah! But if the last atom is blue ...
                tempr = [ x, y, lista[-1][2] + 1, True]       # Make the next red atom
                tempb = [ x + 1, y, lista[-1][2] + 1, False]  # Make the next blue atom
                if tempr[2] <= L:                             # If the Z coordinate is reasonable
                    lista.append( tempr )                     # Add the red coordinates to the list
                    lista.append( tempb )                     # Add the blue coordinates to the list


        restart = [0, y + 1, y, False]                        # Create the new coordinates for the atom on top
        if restart[1] <= L: lista.append(restart)             # And if the Y value is reasonable, add to list 
print("Coordinates: ")
print("#  [x, y, z, color]")
for i, item in enumerate(lista):                      # Adding the carbon number per coordinates
    print(f"{i+1}. {item}")                           # Formatting the printed data

atomos = len(lista)                                   # Count the number of atoms in the list by using its coordinates
print("Number of atoms:", atomos)
# Make this a function >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>




# Make this a function -----------------------------------------------------------------------------------------------------------------------------
lambdas = [0,0,0]

print("-----------------------------------------------------------------")
print("If you want to remove 1/2/3 triangles from the triangulene" )
print('                        C -- C                            ')
print('                       /      \                           ')
print('                      C        C -- C                     ')
print('                       \      /      \                    ')
print('                        C -- C        C                   ')
print('                       /      \      /                    ')
print('                      C        C -- C                     ')
print('                       \      /                           ')
print('                        C -- C                            ')
print('                                                          ')
print("-----------------------------------------------------------------")

print("Please write the size of triangles to be removed following the next formula:")
print("                         ( l1, l2, l3 )                                        ")
print("       l1 = side for the first equilateral triangle                            ")
print("       l2 = side for the second equilateral triangle                           ")
print("       l3 = side for the third equilateral triangle                            ")
print("                                                                               ")
print("             If you don't want a removal write 0                               ")
print("------------------------------------------------------------------")
    
l1 = int(input("Enter value for l1: "))
l2 = int(input("Enter value for l2: "))
l3 = int(input("Enter value for l3: "))

lambdas.sort(reverse=True)
# Removing the first triangle, its coordinates are at the inferior tip from the left side of the triangulene. We have thre conditions in order to eliminate the tips of the triangulene.
# looking at the tringulene, each side of it has three coordinates: x, y and z. Each of them have the coordinates and subsequently a relation between them in order to describe small 
# triangles. The first triangle to be eliminated is at the left bottom corner, from it all the coordinates contained in the "y" axis are going to be the same (L). The second triangle 
# have all its "x" coordinates equal to L and finally the top triangle is going to have all its "z" coordinates in zero. Taking this as a basis the following patterns were described and
# generalized in order to obtain obtain the triangle's coordinates and remove them as the user wish.
L_list = list(range(0, L+1))                                                    # Convert the L number into a list

if l1 > 0:                                                                      # If the user selects the option to remove atoms from l1   
    last_values = L_list[-l1:]                                                  # To select the last l1 numbers contained in the value proportioned by the user "l1" and converting it into a list
    lines_to_remove = [i for i, l in enumerate(lista) if l[1] in last_values]   # Conditions to find those numbers which are going to be eliminated from the "y" axis
    lista = [l for i, l in enumerate(lista) if i not in lines_to_remove]        # To remove all of the line's number contained in "lines_to_remove"

if l2 > 0:                                                                      # If the user selects the option to remove atoms from l2
    last_values2 = L_list[-l2:]                                                 # To select the last l2 numbers contained in the value proportioned by the user "l2" and converting it to a list
    lines_to_remove2 = [i for i, l in enumerate(lista) if l[0] in last_values2] # Conditions to find those numbers which are going to be eliminated from the "x" axis
    lista = [l for i, l in enumerate(lista) if i not in lines_to_remove2]       # To remove all of the line's number contained in "lines_to_remove2"

if l3 > 0:                                                                      # If the user selects the option to remove atoms from l3        
    first_values = L_list[:l3]                                                  # To select the last l3 numbers contained in the value proportioned by the user "l3" and converting it to a list
    lines_to_remove3 = [i for i, l in enumerate(lista) if l[2] in first_values] # Conditions to find those numbers which are going to be eliminated from the "z" axis
    lista = [l for i, l in enumerate(lista) if i not in lines_to_remove3]       # To remove all of the line's number contained in "lines_to_remove3"
    for i, item in enumerate(lista):
      print(f"{i+1}. {item}")

  # If the triangulene have L = even number and the user wants to eliminate three equal triangles, one must reassure that all the selected atoms are removed. In this case this condition is
  # created, which means that if the user select a triangle's side l1=l2=l3, then these are going to share frontiers. This could produce a mistake, because one could not be errasing these frontier
  # atoms. That is why only for L= even number and only if l1=l2=l3 are the same and exactly half of L value, then:
if l1 == l2 == l3 and L % 2 == 0:                                               # the three sides are equal and L is an even number
    conditions = {0, L//2}                                                      # If the coordinates have zero or half of L value
    lines_to_remove_even = [i for i, l in enumerate(lista) if l[0] in conditions and l[1] in conditions and l[3] == True] # Search the final list and look in the "x" and "y" axis if you find the numbers in "conditions" and look if these numbers have in the "z" axis the red color (True)

    lista = [l for i, l in enumerate(lista) if i not in lines_to_remove_even]  # Save the new list not taking in count the values contained in "lines_to_remove_even"
    print("Coordinates after removing triangles: ")
    print("# [x, y, z, color]:")
    for i, item in enumerate(lista):
      print(f"{i+1}. {item}")



# Now all the coordinates belonging to the atoms in the tips of the triangle are eliminated, but there are some atoms that are remaining
# These ones are single red atoms placed exactly at the border bet
#        lines_to_remove = [L - i for i in range(lambdas[2])]    # To convert the line numbers into indices in the list
#        lista = [l for l in lista if not (l[0] in lines_to_remove)]

atomos = len(lista)                                           # Count the number of atoms in the list by using its coordinates
print("New number of atoms:", atomos)
    
#for i, item in enumerate(lista):                             # Adding the carbon number per coordinates
#    print(f"{i+1}. {item}")                                  # Formatting the printed data

# Make this a function -----------------------------------------------------------------------------------------------------------------------------




                                                         # If you say "no", then you don't want a removal
#  l1, l2, l3 = 0, 0, 0                                        # Then l1, l2 and l3 are going to be zero
# Make this a function +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
huckel = np.zeros((atomos,atomos), dtype=np.int32)          # Create integer matrix filled with zeroes
for i, a in enumerate(lista):                               # Creating the conditions for the pattern
      huckel[i,i] = 0                                         # Setting the diagonal elements of the matrix to zero
      if a[3]:                                                # If the atom is a high vertex...
          left = [ a[0], a[1] + 1, a[2], False ]              # Creating the conditions of the pattern
          right = [ a[0] + 1, a[1], a[2], False ]             # Creating the conditions of the pattern
          if left in lista:                                   # If the condition follows the left pattern...
              huckel[i, lista.index(left)] = 1                # Set the corresponding off-diagonal matrix elements to 1
          if right in lista:                                  # If the condition follows the right pattern...
              huckel[i, lista.index(right)] = 1               # Set the corresponding off-diagonal matrix elements to 1
      else:                                                   # If the atom is a low vertex...
          low = [ a[0], a[1], a[2] + 1, True ]                # The atom must be in this finall pattern and must be red
          if low in lista:                                    # If it follows this condition then:
              huckel[i, lista.index(low)] = 1                 # Set the corresponding off-diagonal matrix elements to 1
      
huckel = huckel + huckel.transpose()                        # Calculating the transpose of the array over its main diagonal
print("Matrix:\n")
for l in huckel:
      for c in l:
          print(c, end="\t")
      print("")      
# Make this a function +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++


valores, vectores = np.linalg.eigh(huckel)                    # Compute the eigenvalues and eigenvectors of the matrix
        
print("Eigenvalues:\n" + str(valores))
print("Eigenvectors:\n" + str(vectores))

plt.plot(np.linspace(0, 1, len(valores)), valores, marker='o', label=f"({L}|{l1},{l2},{l3})") # Plot the eigenvalues with an scpecific markpoints and label

plt.legend()                                                  # Adds a legend to the plot with the labels provided
plt.show()                                                    # Display the plot in a new window
```

# Coordinates visualizer

This code helps to see graphically the coordinates in the triangulene.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

print('----------------------------------------------------------')
print('           HÜCKEL HAMILTONIAN VISUALIZER                  ')
print('                 TCCM Winter School                       ')
print('----------------------------------------------------------')
n = int(input("Provide the number of hexagons per side: "))

ax = plt.gca()

l = 0.6
x = l*np.cos(np.pi / 6)
y = l*np.sin(np.pi / 6)

# Axes
plt.plot([0, 1.7*n*x], [0,1.7*n*y], 'k.', linestyle="--")
plt.plot([0,-1.7*n*x], [0,1.7*n*y], 'k.', linestyle="--")
plt.text(-1.7*n*x, 1.7*n*y+0.5, "x")
plt.text( 1.7*n*x, 1.7*n*y+0.5, "y")

plt.text(-0.25*x,-0.6*y, "(0,0,0)")

# Coordinates
coords_red = []
coords_blue = []

coords_red.append([0,0])

for i in range(n): 

    plt.plot([ x*(i + 1),  x*(i + 1)], [y*(i + 1), y + l + i*(y + l)], 'k.', linestyle="--")
    plt.plot([-x*(i + 1), -x*(i + 1)], [y*(i + 1), y + l + i*(y + l)], 'k.', linestyle="--")

    plt.text(-x*(i + 1)-x*0.4, y*(i + 1)-0.2, f"{i+1}")
    plt.text( x*(i + 1)+0.2*x, y*(i + 1)-0.2, f"{i+1}")

    for j in range(i + 1):
        # Base
        red_square = [  [x*(2*j - i), i*(y + l)],
                        [x*(2*j + 1 - i), y+i*(y + l)],
                        [x*(2*j - i), 2*y+i*(y + l)],
                        [x*(2*j - 1 - i), y+i*(y + l)]]
        pts = np.array(red_square)
        p = Polygon(pts, closed=False, fc='#0066cc', ec='none')
        ax.add_patch(p)

        # Lateral wall righ (dark)
        red_square2 = [ [x*(2*j+1-i), y + i*(y + l)],
                        [x*(2*j+1-i), y + l + i*(y + l)],
                        [x*(2*j-i), 2*y + l + i*(y + l)],
                        [x*(2*j-i), 2*y + i*(y + l)]]
        pts = np.array(red_square2)
        p = Polygon(pts, closed=False, fc='#005566', ec='none')
        ax.add_patch(p)

        # Lateral wall left (light)
        red_square3 = [ [x*(2*j-1-i), y + i*(y + l)],
                        [x*(2*j-1-i), y + l + i*(y + l)],
                        [x*(2*j-i), 2*y + l + i*(y + l)],
                        [x*(2*j-i), 2*y + i*(y + l)]]
        pts = np.array(red_square3)
        p = Polygon(pts, closed=False, fc='#006699', ec='none')
        ax.add_patch(p)

    # Show edge coordinates
    plt.text(-(i+1)*x+0.2*x,(i+1)*(y+l) - l, f"({i+1},0,{i})", color="white")
    plt.text(-(i+1)*x+0.2*x,(i+1)*(y+l), f"({i+1},0,{i+1})", color="white")

    # Load coordinates for red-blue dots
    coords_blue.append([-(i+1)*x,(i+1)*(y+l) - l])
    coords_red.append([-(i+1)*x,(i+1)*(y+l)])

# Re-arrange data
red_x, red_y = zip(*coords_red)
blue_x, blue_y = zip(*coords_blue)

# Plot red and blue dots
plt.plot(red_x, red_y, "ro")
plt.plot(blue_x, blue_y, "bo")

ax.set_xlim(-4,4)
ax.set_ylim(-2, 6)

plt.rcParams["figure.figsize"] = (9,9)
plt.rcParams['figure.dpi'] = 100
plt.axis('off')
plt.show()
```
